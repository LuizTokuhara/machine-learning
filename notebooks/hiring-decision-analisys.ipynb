```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('../dataset/recruitment_data.csv', sep=',', decimal='.')
df.info()
```

```python
RS_Counts=df['RecruitmentStrategy'].value_counts().sort_values()
labels={1: 'Agressivo',
    2: 'Moderado',
    3: 'Conservador'}
colors=['lightblue','yellow','lightpink']
plt.pie(x=RS_Counts,labels = [labels[i] for i in RS_Counts.index],autopct='%1.1f%%',colors=colors)
plt.axis('equal')
plt.title('Estrat√©gia de recrutamento')
plt.show()
```

```python
# Filter data to include only hired candidates (HiringDecision == 1)
hired_df = df[df['HiringDecision'] == 1]

# Calculate the percentage of hired candidates by recruitment strategy
strategy_percentages = hired_df.groupby('RecruitmentStrategy')['HiringDecision'].count() / len(hired_df) * 100

# Define custom labels for the recruitment strategies
strategy_labels = ['Agressivo', 'Moderado', 'Conservador']

# Create a bar plot
plt.figure(figsize=(8, 8))
sb.barplot(x=strategy_percentages.index, y=strategy_percentages.values, hue=strategy_percentages.index, palette='Paired')

# Add labels
plt.ylabel('Porcentagem de candidados contratados')
plt.title('Porcentagem de candidados contratados por estrat√©gia')
plt.xticks(ticks=range(len(strategy_labels)), labels=strategy_labels, rotation=45)
plt.tight_layout()

plt.show()
```

```python
hired_df = df[df['HiringDecision'] == 1]

strategy_percentages = hired_df.groupby('EducationLevel')['HiringDecision'].count() / len(hired_df) * 100

education_labels = ['Bacharelado 1', 'Bacharelado 2', 'Mestrado', 'PhD']

plt.figure(figsize=(8, 8))
sb.barplot(x=strategy_percentages.index, y=strategy_percentages.values, hue=strategy_percentages.index, palette='Paired')

# Add labels
plt.ylabel('Porcentagem de candidados contratados')
plt.title('Porcentagem de candidados contratados por n√≠vel de escolaridade')
plt.xticks(ticks=range(len(education_labels)), labels=education_labels, rotation=45)
plt.tight_layout()

plt.show()
```

```python
hired_df = df[df['HiringDecision'] == 1]

strategy_percentages = hired_df.groupby('ExperienceYears')['HiringDecision'].count() / len(hired_df) * 100

plt.figure(figsize=(8, 8))
sb.barplot(x=strategy_percentages.index, y=strategy_percentages.values, hue=strategy_percentages.index, palette='Paired')

plt.ylabel('Porcentagem de candidados contratados')
plt.title('Porcentagem de candidados contratados por anos de experi√™ncia')

plt.tight_layout()

plt.show()
```

```python
hired_df = df[df['HiringDecision'] == 1]

strategy_percentages = hired_df.groupby('PreviousCompanies')['HiringDecision'].count() / len(hired_df) * 100

plt.figure(figsize=(8, 8))
sb.barplot(x=strategy_percentages.index, y=strategy_percentages.values, hue=strategy_percentages.index, palette='Paired')

plt.ylabel('Porcentagem de candidados contratados')
plt.title('Porcentagem de candidados contratados por n√∫mero de empresas')

plt.tight_layout()

plt.show()
```
